# CLEANING DATA

```python
df.isnull().sum()
```

```python
# hapus row yang ada data null
df.dropna(inplace=True)
```

```python
# ngisi dengan angka tertentu (0, median, mean)
df['country'].filna(0, inplace=True)
df['kolom'].filna(df['kolom'].median(), inplace=True)
df['kolom'].filna(df['kolom'].mean(), inplace=True)
```

```python
df.columns = df.columns.str.replace(' ', '_').str.lower()
```

```python
df.info()
```

```python
# konversi data type data
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
```

```python
df.info()
```

# ANALISIS DATA


```python
df.groupby('category')['sales'].sum().sort_values(ascending=False)
```

```python
df.groupby('category')['profit'].sum().sort_values(ascending=False)
```

```python
sns.barplot(x='category', y='sales', data=df, estimator=sum, color='pink', errorbar=None)
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title('Correlation Matrix')
plt.show()
```
